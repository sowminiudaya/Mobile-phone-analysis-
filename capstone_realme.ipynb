```python

# List of URLs
links = [
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy,4io'
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=2',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=3',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=4',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=5',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=6',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=7',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=8',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=9',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=10',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=11',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=12',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=13',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=14',
    'https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy%2C4io&page=15'
   
    ]

# List to store scraped data
data = []

# Iterate over each URL
for link in links:
    try:
        # Make a request to each URL
        response = requests.get(link, headers={"User-Agent": "Mozilla/5.0"})
        response.raise_for_status()  # Check if the request was successful
        
        # Parse the HTML response
        sel = Selector(text=response.text)
        
        # Extract data
        product_names = sel.css("div.KzDlHZ::text").getall()
        prices = sel.css("div.Nx9bqj._4b5DiR::text").getall()
        discounts = sel.css("div.UkUFwK > span::text").getall()
        exchange_offers = sel.css("div:nth-child(4) > div > div:nth-child(3)::text").getall()
        ram_rom = sel.css("ul > li:nth-child(1)::text").getall()
        battery = sel.css("ul > li:nth-child(4)::text").getall()
        display = sel.css("ul > li:nth-child(2)::text").getall()
        ratings = sel.css("div > div > div > a > div.yKfJKb.row > div.col.col-7-12 > div._5OesEi > span.Wphh3N > span > span:nth-child(1)::text").getall()
        warranty = sel.css("ul > li:nth-child(6)::text").getall()
        reviews = sel.css('div > div > div > a > div.yKfJKb.row > div.col.col-7-12 > div._5OesEi > span.Wphh3N > span > span:nth-child(3)::text').getall()
        camera = sel.css("div > div > div > a > div.yKfJKb.row > div.col.col-7-12 > div._6NESgJ > ul > li:nth-child(3)::text").getall()
        actual_price = sel.css("div > div > div > a > div.yKfJKb.row > div.col.col-5-12.BfVC2z > div.cN1yYO > div.hl05eU > div.yRaY8j.ZYYwLA::text").getall()
        processor = sel.css("div > div > div > a > div.yKfJKb.row > div.col.col-7-12 > div._6NESgJ > ul > li:nth-child(4)::text").getall()

        actual_price = [re.sub(r'₹', '', price) for price in actual_price]
        actual_price = list(filter(None, actual_price))

        # Store data in list
        for i in range(len(product_names)):
            data.append({
                "Product Name": product_names[i] if i < len(product_names) else None,
                "Price": prices[i] if i < len(prices) else None,
                "Discount": discounts[i] if i < len(discounts) else None,
                "Exchange Offer": exchange_offers[i] if i < len(exchange_offers) else None,
                "RAM/ROM": ram_rom[i] if i < len(ram_rom) else None,
                "Battery": battery[i] if i < len(battery) else None,
                "Display": display[i] if i < len(display) else None,
                "Ratings": ratings[i] if i < len(ratings) else None,
                "Warranty": warranty[i] if i < len(warranty) else None,
                "Reviews": reviews[i] if i < len(reviews) else None,
                "Camera": camera[i] if i < len(camera) else None,
                "Actual Price": actual_price[i] if i < len(actual_price) else None,
                "Processor": processor[i] if i < len(processor) else None
            })
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {link}: {e}")

# Convert data to DataFrame and save to CSV
if data:
    df12 = pd.DataFrame(data)
    df12.to_csv("I_Phone.csv", index=False)


df12
    
```
