```python
import pandas as pd
import re

# Load the dataset
file_path = "android_mobiles.csv"
df = pd.read_csv(file_path)

# Function to clean specific columns
def clean_text(text, pattern):
    if pd.isna(text):  # Handle NaN values
        return text
    return re.sub(pattern, '', text).strip()

# Apply cleaning functions
df['Price'] = df['Price'].apply(lambda x: clean_text(x, r'₹'))  # Remove ₹ symbol
df['Discount'] = df['Discount'].apply(lambda x: clean_text(x, r'% off'))  # Remove % off
df['Ratings'] = df['Ratings'].apply(lambda x: clean_text(x, r' Ratings'))  # Remove ' Ratings' text
df['Reviews'] = df['Reviews'].apply(lambda x: clean_text(x, r' Reviews'))  # Remove ' Reviews' text

df
```

```python
def extract_ram(text):
    match = re.search(r'(\d+) GB RAM', str(text))
    return match.group(1) if match else None

def extract_rom(text):
    match = re.search(r'(\d+) GB ROM', str(text))
    return match.group(1) if match else None

df['RAM'] = df['RAM/ROM'].apply(extract_ram)
df['ROM'] = df['RAM/ROM'].apply(extract_rom)

df.drop(columns=['RAM/ROM'], inplace=True)  # Remove the original RAM/ROM column

df
```

```python
df = df.drop_duplicates()
df = df.drop(columns=['Exchange Offer', 'Warranty'])

df
```

```python
df.dropna(inplace=True)
```
